```python
#adding title of x and y axis
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x =  np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('x**')
plt.title('Parabola')
plt.show()
```

## Bar Plot

```python
bar_graph = plt.figure()
ax = bar_graph.add_axes([0, 0, 1, 2])
languages = ['English', 'French', 'Italian', 'Spanish', 'Russian']
people = [100, 50, 30, 55, 20]
ax.bar(languages, people)
plt.xlabel('Languages')
plt.ylabel('No of people')
plt.show()
```

```python
ax = plt.figure().add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Italian', 'Spanish', 'Russian']
people = [100, 50, 30, 55, 20]
ax.bar(languages, people, color = 'g')
plt.xlabel('Languages')
plt.ylabel('No of people')
plt.show()
```

PIE CHART

```python
my_pie = plt.figure().add_axes([0, 0, 1, 1])
languages = ['English', 'French', 'Italian', 'Spanish', 'Russian']
people = [100, 50, 30, 55, 20]
my_pie.pie(people, labels = languages, autopct = '%1.1f%%')
plt.title('Languages used')
plt.show()
```

## Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x) 
fig2 = plt.figure().add_axes([0, 0, 1, 1])
fig2.scatter(x, y, color = 'y')
fig2.scatter(x, z, color = 'b')
```

```python
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z) 
fig2 = plt.axes(projection = '3d')
fig2.scatter(x, y, z, c = z, cmap = 'Reds')
```

```python
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z) 
fig2 = plt.axes(projection = '3d')
fig2.scatter(x, y, z, c = z, cmap = 'Greens')
```

```python
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z) 
fig2 = plt.axes(projection = '3d')
fig2.scatter(x, y, z, c = z, cmap = 'Blues')
```
